```python
# Sorting player names by their scores 

df.sort_values(by=['Goals'], inplace=False, ascending=False)
```

```python
# Sorting player names with index of club and position by their scores 

PlayerGoals = df.groupby(['Name', 'Club','Position'])[["Goals"]].sum()
PlayerGoalsSorted = PlayerGoals.sort_values(['Goals'], ascending = False)
PlayerGoalsSorted
```

```python
# Remove entries which nationality 

df = df[df['Nationality'].notna()]

# Visualize number of players appearances with thier position and nationality

fig = px.sunburst(df, 
                   path=['Position', 'Nationality'], 
                   values='Appearances', 
                 )
fig.update_layout(title_text='<b>Players position by country <b>', 
                  title_x=0.5, 
                  titlefont=dict(color='black', 
                            size=28, 
                            family="Courier New, monospace",),
                  width=600,
                  height=600,
                  showlegend=False,
                 )
iplot(fig)
```

```python
# Big chances missed overall

ChancesMissed = df.groupby(['Name'])[['Big chances missed']].sum()
ChancesMissed = ChancesMissed.sort_values(['Big chances missed'], ascending = False)[:5]


ChancesMissed
```

```python
# Big chances missed per match

# take care off the data type for division
datta_appNonZero = df[df['Appearances'] > 0]
features = df.columns
df1 = df[features]

# separate cols with dtype objects and cols that may not be divided by appearances
cols = features.drop(['Age', 'Name', 'Appearances', 'Club', 'Nationality', 'Jersey Number', 'Cross accuracy %', 'Position', 'Goals per match', 
                      'Passes per match','Tackle success %', 'Shooting accuracy %'])
datta_appNonZero.loc[:, cols] = datta_appNonZero.loc[:, cols].div(datta_appNonZero['Appearances'], axis=0)

# palyers who have made atleast 38 games (a seasons worth games)
# data normalized
all_players = datta_appNonZero[datta_appNonZero['Appearances'] >= 38]

ChancesMissedPergame = all_players.sort_values('Big chances missed', ascending=False).head(5)
ChancesMissedPergame = ChancesMissedPergame.groupby(['Name'])[['Big chances missed']].sum()


ChancesMissedPergame
```
